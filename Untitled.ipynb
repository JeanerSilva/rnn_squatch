```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image, display

class BasicNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.w00 = nn.Parameter(torch.tensor(1.7), requires_grad=False)
        self.b00 = nn.Parameter(torch.tensor(-0.85), requires_grad=False)
        self.w01 = nn.Parameter(torch.tensor(-40.8), requires_grad=False)
        
        self.w10 = nn.Parameter(torch.tensor(12.6), requires_grad=False)
        self.b10 = nn.Parameter(torch.tensor(0.0), requires_grad=False)
        self.w11 = nn.Parameter(torch.tensor(2.7), requires_grad=False)
        
        self.final_bias = nn.Parameter(torch.tensor(0.), requires_grad=True)
        
    def forward(self, input):
        input_to_top_relu = input * self.w00 + self.b00
        top_relu_output = F.relu(input_to_top_relu)
        scaled_top_relu_output = top_relu_output * self.w01

        input_to_bottom_relu = input * self.w10 + self.b10
        bottom_relu_output = F.relu(input_to_bottom_relu)
        scaled_bottom_relu_output = bottom_relu_output * self.w11

        input_to_final_relu = scaled_top_relu_output + scaled_bottom_relu_output + self.final_bias

        output = F.relu(input_to_final_relu)
        return output  

# Criação do modelo
model = BasicNN()

# Criação de dados de entrada e rótulos
inputs = torch.tensor([0., 0.5, 1.], requires_grad=True)
labels = torch.tensor([0., 1., 0.], requires_grad=False)

# Definição do otimizador
optimizer = SGD(model.parameters(), lr=0.1)

final_bias_values = []

# Loop de treinamento
for epoch in range(100):
    total_loss = 0
    for iteration in range(len(inputs)):
        input_i = inputs[iteration]
        label_i = labels[iteration]
        
        output_i = model(input_i)
        
        loss = (output_i - label_i)**2
        
        loss.backward()
        
        total_loss += float(loss)
        
        optimizer.step()
        optimizer.zero_grad()
        
        final_bias_values.append(model.final_bias.data.item())
    
    if total_loss < 0.0001:
        print("Num steps: " + str(epoch))
        break
        
    print("Step: " + str(epoch).zfill(2) + " - Final Bias: " + str(model.final_bias.data))
    
print("Final bias, after optimization: " + str(model.final_bias.data))
```

```python
input_doses = torch.linspace(start=0, end=1, steps=11)
model = BasicNN()
output_values = model(input_doses)

plt.plot(input_doses, output_values.detach(), color='green', linewidth=2.5)
plt.ylabel('Efetividade')
plt.xlabel('Dose')
plt.title('Relação entre Dose e Efetividade')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(final_bias_values, marker='o', color='blue', linewidth=2.5)
plt.ylabel('Final Bias')
plt.xlabel('Epoch')
plt.title('Gradient Descent of Final Bias')
plt.grid(True)
plt.show()
```
